# Data Cleaning

```python
star.head()
```

```python
star['Beverage_category'].value_counts()
```

```python
star['Beverage'].value_counts()
```

```python
star['Beverage_prep'].value_counts()
```

```python
# Renaming columns 
star = star.rename(columns = {'Beverage_category': 'Drink', 'Beverage_prep': 'Preperation'})
```

```python
star.head()
```

```python
# Remove percentage signs 
star.columns = [col.replace('% ', '').strip() for col in star.columns]
```

```python
star.head()
```

```python
star.columns = [col.replace(' ', '_').strip() for col in star.columns]
```

```python
star.head()
```

```python
# Convert types
star.dtypes
```

```python
# Change columns to floats
columns_to_float = [
    'Vitamin_A_(DV)', 'Vitamin_C_(DV)', 'Calcium_(DV)'
]

for col in columns_to_float:
    star[col] = star[col].astype(float)
```

```python
star.dtypes
```

```python
# clean NaN in Caffeine to change it to int
sns.boxplot(data = star, 
            x = 'Caffeine_(mg)')
```

```python
# there are outliers so lets use the mediam to replace NaN
star['Caffeine_(mg)'].median()
```

```python
star['Caffeine_(mg)'] = star['Caffeine_(mg)'].fillna(star['Caffeine_(mg)'].median())
```

```python
star.isna().sum()
```

```python
# Change columns to int64
columns_to_int = [
    'Protein_(g)', 'Total_Fat_(g)', 'Trans_Fat_(g)', 'Saturated_Fat_(g)', 'Caffeine_(mg)'
]

for col in columns_to_int:
    star[col] = star[col].astype(int)
```
